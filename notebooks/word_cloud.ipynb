```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import wordcloud
from wordcloud import WordCloud, ImageColorGenerator 
import requests
import numpy as np
from PIL import Image
import os
import urllib
import cv2
```

```python
os.getcwd()
```

```python
with open('../data/text/positive.txt') as f:
    pos_lines = f.readlines()

with open('../data/text/negative.txt') as f:
    neg_lines = f.readlines()
```

```python
pos_words = pd.DataFrame(data=pos_lines, columns=['words'])
neg_words = pd.DataFrame(data=neg_lines, columns=['words'])
```

```python
neg_words.head()
```

```python
pos_words.head()
```

```python
pos_words['words'] = pos_words['words'].str.replace('\n', '')
neg_words['words'] = neg_words['words'].str.replace('\n', '')
```

```python
pos_words.head()
```

```python
negative_words=[]
positive_words=[]

for i in pos_words['words']:
    positive_words.append(i)
    
for i in neg_words['words']:
    negative_words.append(i)
```

```python
mask = np.array(Image.open("../data/masks/upvote.png"))

wordcloud = WordCloud(collocations=False,
                      mask=mask, 
                      background_color='white', 
                      min_font_size=10, 
                      height=600, 
                      width=1000,
                      #contour_color='#000000' 
                     ).generate(' '.join(positive_words))

image_colors = ImageColorGenerator(mask)
wordcloud.recolor(color_func=image_colors)


plt.figure(figsize=(30,20))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
# store to file
plt.savefig("cloud.png", format="png")
plt.show()
```
